```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10)
y = 20 + (x ** 2)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-10,10)
plt.ylim(0,110)
plt.title('X vs Y');
```

```python
fig = plt.figure(figsize=(5,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```
